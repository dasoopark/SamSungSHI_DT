```python
# 1이상의 정수 n을 입력받은후 1 ~ N까지의 합을 구하는 함수 작성 
def total(a):
    sum = 0
    for i in range(1,a+1):
        sum+= i
    return sum
print(total(100))
```

```python
#숫자 요소를 가지는 리스트를 입력받은 후 합과 평균을 구하고 튜플 형태로 반환하는 함수 작성
def calculator(a):
    sum = 0
    average = 0
    for i in a:
        sum += i
    average = sum / len(a)
    result = [sum,average]
    return tuple(result)

print(calculator([3,7,9,11]))
```

```python
#아래와 같은 수열이 있을 때 N번째 수열의 값을 구하는 함수 작성

def 수열(a):
    temp = 0
    total = 0
    for i in range(1,a+1):
        temp = temp+i
        total = total+temp
    return total
수열(4)
```

```python
#문자열의 가운데 글자를 반환하는 함수 작성
def 가운데(a):
    중간 = len(a)%2
    가운데글자 = len(a)//2
    if 중간 == 0:
        return a[가운데글자-1:가운데글자+1]
    else:
        return a[가운데글자]
print(가운데('functions'))
print(가운데('python'))
```

```python
#길이가 N이고, 수박수박수박수...와 같은 패턴을 반환하는 함수 작성
# 입력 예) 출력 예)
# n=3 수박수
# n=4 수박수박
# n=7 수박수박수박수
# n=11 수박수박수박수박수박수
def 수박(a):
    보관 = ['수','박']
    수 = 0
    박 = 0
    temp = 0
    for i in range(1,a+1):
        if i%2==0:
            박 = 1
            temp = 박
        elif i%2==1:
            수 = 0
            temp = 수
        print(보관[temp],end='')
            
            
수박(11)
```

```python
# 자연수 N이 주어지면 N의 각 자리 수 합을 구하는 함수 작성
def 자연수(n):
    temp = n
    temp2 = n
    total = 0
    while True:
        if(temp==0):
            break
        temp2 = temp%10
        total+=temp2
        temp = temp//10
    return total

print(자연수(13579))
```

```python
#홀수 마방진
# 규칙 1) 다음 숫자는 오른쪽 상단으로 이동

# 규칙 2) 범위(위 또는 오른쪽)를 벗어나면
#              반대쪽(아래 또는 왼쪽)으로 이동

# 규칙 3) 이동하려는 위치에 0이 아닌 값이
#              있으면 이동 전 위치의 아래로 이동

m = [ [0 for col in range(5)] for row in range(5) ]

row = 0
col = 2
m[row][col] = 1 #(0,2)위치에 1 입력
for i in range(2, 26): # 2 ~ 25 1번 =>(0,2)에 들어감
    x = row 
    y = col
    row -= 1
    col += 1
    
    if row < 0:
        row = 4
    if col > 4:
        col = 0
    if m[row][col] !=0:
        row = x+1
        col = y
    
    m[row][col] = i
```

```python
# 1이 될 때까지
# 어떠한 수 N이 될 때까지 다음의 두 과정 중 하나를 반복적으로 선택하여 수행하려고 합
# 니다. 단, 두번째 연산은 N이 K로 나누어 떨어질 때만 선택할 수 있습니다.
# 1. N에서 1을 뺍니다.
# 2. N을 K로 나눕니다.
# 예를 들어 N이 17, K가 4라고 가정합시다. 이 때 1번의 과정을 한 번 수행하면 N은 16이
# 됩니다. 이후에 2번의 과정을 두 번 수행하면 N은 1이 됩니다. 결과적으로 이 경우 전체
# 과정을 실행한 횟수는 3이 됩니다. 이는 N을 1로 만드는 최소 횟수입니다.
# N과 K가 주어질 때 N이 1이 될 때까지 1번 혹은 2번의 과정을 수행해야 하는 최소 횟수
# 를 구하는 프로그램을 작성하세요.


def sol(n,k):
    cnt = 0
    while n!=1:
        if n % k == 0:
            n //= k
            cnt+=1
        else:
            n -= 1
            cnt+=1
    return cnt
print(sol(25,5))
```
